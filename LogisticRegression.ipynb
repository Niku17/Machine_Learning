```python
# Importing the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Importing the dataset.
df=pd.read_csv("Social_Network_Ads.csv")

# Creating matrix of feature and vector of dependent variable.
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values

# Splitting into training and testing dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=1)

# feature scaling on matrix of feature so that none of the feature is ignored
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)

# Training Logistic regression model on dataset
from sklearn.linear_model import LogisticRegression
Lr=LogisticRegression(random_state=0)
Lr.fit(x_train,y_train)

# Now let predict the result for testing set.
y_pred=Lr.predict(x_test)

# Now let compare both predicted and actual values
print(np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1)), axis=1))

# Let check for a single observation i.e age=30years and etimated salary=87000 dollars
print(Lr.predict(sc.transform([[30,87000]])))

# creating confusion network to check the accuracy of the model.
from sklearn.metrics import confusion_matrix,accuracy_score
confusion_matrix(y_pred,y_test)
accuracy_score(y_pred,y_test)




```
